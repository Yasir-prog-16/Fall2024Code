```python
import json
import networkx as nx
import matplotlib.pyplot as plt
import gurobipy as gp
from gurobipy import GRB

# Open and read the JSON file
with open('Red_Optimizers.json', 'r') as file:
    data = json.load(file)

# Create a directed graph
G = nx.DiGraph()

# Adding nodes 
counter = 0
for element in data:
    G.add_node(counter, **element)  
    counter += 1

# Donor-recipient compatibility
donor_recipient_compatibility = {'O': ['O', 'A', 'B', 'AB'], 'A': ['A', 'AB'], 'B': ['B', 'AB'], 'AB': ['AB']}

# Adding edges based on compatibility
for node in G.nodes:
    donors = G.nodes[node]['Donor']  # This is a list of donor blood types
    for donor in donors:  # Iterating over each donor type in the list
        for target in G.nodes:
            if node == target:  # Skipping self-loops
                continue
            recipient = G.nodes[target]['Recipient']
            if recipient in donor_recipient_compatibility[donor]:
                G.add_edge(node, target)

# Display edges for verification
#print('Edges:', list(G.edges))

# Finding cycles of size 2 and 3
# Cycles of size 3
cycle_3 = []
H = G.to_undirected()

for (i, j) in G.edges:
    for k in nx.common_neighbors(H, i, j):
        if {i, j, k} in cycle_3:
            continue
        if (i, j) in G.edges and (j, k) in G.edges and (k, i) in G.edges:
            cycle_3.append((i, j, k))

print("Cycles of length 3:", cycle_3)

# Cycles of size 2
cycle_2 = []
for (i, j) in G.edges:
    if (j, i) in G.edges and (j, i) not in cycle_2:
        cycle_2.append((i, j, i))

print("Cycles of length 2:", cycle_2)

# Combining all cycles
cycles = cycle_2 + cycle_3

# Optimize Using Gurobi
# Create Gurobi model
m = gp.Model("Kidney_Paired_Donation")

# Decision variables:
x = m.addVars(cycles, vtype=GRB.BINARY, name="cycle")

# Objective: maximize the number of transplants
m.setObjective(gp.quicksum(x[c] for c in cycles), GRB.MAXIMIZE)

# Constraint: Each node can participate in at most one cycle
for node in G.nodes:
    m.addConstr(
        gp.quicksum(x[c] for c in cycles if node in c) <= 1,
        name=f"node_{node}_participation"
    )

# Solve the model
m.optimize()

# Displaying the Results
if m.status == GRB.OPTIMAL:
    print(f"Optimal number of transplants: {int(m.objVal)}")
    selected_cycles = [c for c in cycles if x[c].x > 0.5]
    print("Selected cycles:")
    for cycle in selected_cycles:
        print(cycle)
else:
    print("No optimal solution found.")

# # Visualize the Graph 
# pos = nx.spring_layout(G)
# nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500)
# nx.draw_networkx_labels(G, pos, labels={node: f"{node}" for node in G.nodes})
# plt.title("Kidney Paired Donation Graph")
# plt.show()
```

```python
from collections import Counter

print("\nCollecting donor-recipient pairs from selected cycles...")
donor_recipient_pairs = []
pair_counts_dynamic = Counter()  # Initialize a Counter for dynamic counting

for cycle in selected_cycles:
    for i in range(len(cycle)):
        donor_node = cycle[i]
        recipient_node = cycle[(i + 1) % len(cycle)]  # Next node in the cycle

        # Skip if donor and recipient are the same node (same family)
        if donor_node == recipient_node:
            print(f"Skipping invalid donor-recipient pair: Donor Node {donor_node} is the same as Recipient Node {recipient_node}.")
            continue

        donor_blood_types = G.nodes[donor_node]['Donor']
        recipient_blood_type = G.nodes[recipient_node]['Recipient']
        print(f"Cycle: {cycle}, Donor Node: {donor_node}, Recipient Node: {recipient_node}")
        print(f"Donor Blood Types: {donor_blood_types}, Recipient Blood Type: {recipient_blood_type}")

        for donor_blood_type in donor_blood_types:
            if donor_blood_type == 'AB':  # Skip if the donor is AB
                print(f"Skipping donor-recipient pair ({donor_blood_type} -> {recipient_blood_type}) due to donor AB.")
                continue
            donor_recipient_pairs.append((donor_blood_type, recipient_blood_type))
            pair_counts_dynamic[(donor_blood_type, recipient_blood_type)] += 1  # Update the count
            print(f"Added donor-recipient pair: ({donor_blood_type}, {recipient_blood_type})")
            print(f"Current Count for ({donor_blood_type} -> {recipient_blood_type}): {pair_counts_dynamic[(donor_blood_type, recipient_blood_type)]}")

# Print final counts after processing all cycles
print("\nFinal Counts of Donor-Recipient Pairs:")
for pair, count in pair_counts_dynamic.items():
    print(f"Pair {pair}: Count = {count}")
```

```python
import pandas as pd
# Count occurrences of each donor-recipient pair
print("\nCounting occurrences of each donor-recipient pair...")
pair_counts = pd.Series(donor_recipient_pairs).value_counts().reset_index()
pair_counts.columns = ['Donor-Recipient Pair', 'Count']

# Split the 'Donor-Recipient Pair' into separate columns
pair_counts[['Donor', 'Recipient']] = pd.DataFrame(pair_counts['Donor-Recipient Pair'].tolist(), index=pair_counts.index)

# Calculate total count of all pairs
total_pairs = pair_counts['Count'].sum()
print(f"\nTotal donor-recipient pairs: {total_pairs}")

# Calculate the proportion for each donor-recipient pair
pair_counts['Proportion'] = pair_counts['Count'] / total_pairs

# Drop the 'Donor-Recipient Pair' column
pair_counts = pair_counts.drop(columns=['Donor-Recipient Pair'])

# Print the final table of donor-recipient pairs, their counts, and proportions
print("\nFinal Donor-Recipient Pair Proportions:")
print(pair_counts)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Example: Final pair_counts DataFrame (you should use the one generated in your code)
# Assuming 'pair_counts' has columns: ['Donor', 'Recipient', 'Count', 'Proportion']

# Plotting the bar chart
plt.figure(figsize=(10, 6))

# Plot for each donor blood type
for donor in pair_counts['Donor'].unique():
    subset = pair_counts[pair_counts['Donor'] == donor]
    plt.bar(
        subset['Recipient'], 
        subset['Proportion'], 
        label=f"Donor {donor}", 
        alpha=0.7
    )

plt.title("Proportions of Donor-Recipient Pairs")
plt.xlabel("Recipient Blood Type")
plt.ylabel("Proportion")
plt.legend(title="Donor Blood Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
